```python
import os
from skimage.io import imread, imsave
from scipy.ndimage import median_filter

def apply_median_filter_to_tiffs(base_folder):
    # Get all folders in base_folder that start with "JHH"
    fov_folders = [f for f in os.listdir(base_folder) if f.startswith("JHH") and os.path.isdir(os.path.join(base_folder, f))]

    # Iterate through each JHH folder
    for fov in fov_folders:
        fov_path = os.path.join(base_folder, fov)
        filtered_fov_path = os.path.join(base_folder, f"{fov}_filtered")

        # Create filtered folder if it doesn't exist
        os.makedirs(filtered_fov_path, exist_ok=True)

        # Process each TIFF file in the folder
        for file in os.listdir(fov_path):
            if file.endswith(".tiff") or file.endswith(".tif"):
                file_path = os.path.join(fov_path, file)

                # Read image
                image = imread(file_path)

                # Apply median filtering
                filtered_image = median_filter(image, size=3)

                # Save the filtered image in the new folder with the same name
                filtered_file_path = os.path.join(filtered_fov_path, file)
                imsave(filtered_file_path, filtered_image)
                print(f"Saved filtered image: {filtered_file_path}")
```

```python
# Set the base directory where folders are located & apply function 
base_directory = "../data/Revision_tiffs/image_data"  
apply_median_filter_to_tiffs(base_directory)
```
